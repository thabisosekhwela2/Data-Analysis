## 1.2. Converting String data to int

```python

# converting to Json file

   
############################
import pandas as pd
f = open("Bike_Buyer_Data_edited.txt", 'r')
answer = []
for line in f:
    app = line.split(',')
    answer.append(app)

f.close()

out = answer[0]

maindict = []

for j in range(1,len(answer)):
    innerdict = {}
    for i in range(len(out)-1):
        key = out[i]
        keyvalue = answer[j][i]
        if keyvalue.isnumeric():
            keyvalue = int(keyvalue)
        # store and convert to numeric representation
        else:
            keyvalue = int(''.join(str(ord(c)) for c in keyvalue))
        innerdict[key] = keyvalue 
    maindict.append(innerdict) 
#print(maindict) 

#dataFrame = pd.DataFrame(maindict, index=id_keys)
df = pd.DataFrame(maindict)
df.head()

```

# 2. Machine Learning Models

```python
# using logistic r
#result = df.loc[df['Region'] == 'Pacific']
# Show the dataframe


#print(result)
df = df[['Income', 'Children', 'Cars', 'Commute Distance', 'Age', 'Purchased Bike']]
label = df.iloc[:,5]
data = df.iloc[:,:4]

#print(label)
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data, label , test_size=0.1, random_state=1)

x_train, x_val, y_train, y_val  = train_test_split(x_train, y_train, test_size=0.1, random_state=1) # 0.25 x 0.8 = 0.2
#print(y_train.shape[0])

```

# Desicion Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
print("score on test: "  + str(clf.score(x_test, y_test)))
print("score on train: " + str(clf.score(x_train, y_train)))

# df = df[['Income', 'Children', 'Cars', 'Commute Distance', 'Age', 'Purchased Bike']] 0.1
# # accurary 0.7 df = df[['Income', 'Children', 'Cars', 'Commute Distance', 'Age', 'Purchased Bike']] 0.05 / 0.2
y_pred = clf.predict(x_val)
y_true = numpy.array(y_val)

from sklearn.metrics import confusion_matrix

TN, FP, FN, TP = confusion_matrix(y_true, y_pred).ravel()

#print('True Positive(TP)  = ', TP)
#print('False Positive(FP) = ', FP)
#print('True Negative(TN)  = ', TN)
#print('False Negative(FN) = ', FN)

print('accuracy is :', (TP + TN)/(TP + TN + FP + FN))
print('balanced accuracy is :', 0.5*((TP/(TP + TN)) + (TN/(TN+FP))))
print('precision is :', TP/(TP + FP))
print("recall is :", TP/(TP + FN))
```

## Random Forest Classifier

```python
#print(result)
df = df[['Income', 'Children', 'Cars', 'Commute Distance', 'Age', 'Purchased Bike']]
label = df.iloc[:,5]
data = df.iloc[:,:4]

#print(label)
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data, label , test_size=0.05, random_state=1)

x_train, x_val, y_train, y_val  = train_test_split(x_train, y_train, test_size=0.2, random_state=1) # 0.25 x 0.8 = 0.2




# filter data by cataegory\
# good for feature rich dataset
from sklearn.ensemble import RandomForestClassifier
# n_estimators = number of decision trees
rf = RandomForestClassifier(n_estimators=30, max_depth=9)
rf.fit(x_train, y_train)
print("score on test: " + str(rf.score(x_test, y_test)))
print("score on train: "+ str(rf.score(x_train, y_train)))

# good for many features
# df = df[['Income', 'Children', 'Cars', 'Commute Distance', 'Age', 'Purchased Bike']] 0.1
# df = df[['Income', 'Children', 'Cars', 'Commute Distance', 'Age', 'Purchased Bike']] 0.015
# accurary 0.7 df = df[['Income', 'Children', 'Cars', 'Commute Distance', 'Age', 'Purchased Bike']] 0.1 / 0.1
# accurary 0.7 df = df[['Income', 'Children', 'Cars', 'Commute Distance', 'Age', 'Purchased Bike']] 0.05 / 0.2


y_pred = rf.predict(x_val)
y_true = numpy.array(y_val)


#from sklearn.metrics import plot_confusion_matrix
#conMat = plot_confusion_matrix(rf, x_test, y_test, cmap='Blues', values_format='.3g')



from sklearn.metrics import confusion_matrix

TN, FP, FN, TP = confusion_matrix(y_true, y_pred).ravel()

#print('True Positive(TP)  = ', TP)
#print('False Positive(FP) = ', FP)
#print('True Negative(TN)  = ', TN)
#print('False Negative(FN) = ', FN)

print('accuracy is :', (TP + TN)/(TP + TN + FP + FN))
print('balanced accuracy is :', 0.5*((TP/(TP + TN)) + (TN/(TN+FP))))
print('precision is :', TP/(TP + FP))
print("recall is :", TP/(TP + FN))
```
